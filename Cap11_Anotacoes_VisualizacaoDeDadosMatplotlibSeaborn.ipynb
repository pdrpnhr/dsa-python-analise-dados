#### Gráfico de área empilhada

#### Gráfico de pizza

#### Gráficos de linha

- Verificar a evolução do comportamento de uma variável

```python
from pylab import *
```

```python
# Dados
x = linspace(0,5,10)
y = x ** 2

# Criando a área de plotagem
fig = plt.figure()

# Definindo a escala dos eixos  
axes = fig.add_axes([0,0,0.8,0.8])

# Cria o plot
axes.plot(x, y, 'r')

# Labels e título
axes.set_xlabel('x')
axes.set_ylabel('y')
axes.set_title('Gráfico de Linha')
```

##### 2 gráficos numa mesma figura

```python
# Gráfico de linha com 2 figuras

# Dados
x = linspace(0,5,10)
y = x ** 2

# Cria a figura
fig = plt.figure()

# Cria os eixos
axes1 = fig.add_axes([0.1,0.1,0.8,0.8])
axes2 = fig.add_axes([0.2,0.5,0.4,0.3])

# Figura principal
axes1.plot(x,y,'r')
axes1.set_xlabel('x')
axes1.set_ylabel('y')
axes1.set_title('Figura Principal')

# Figura secundária
axes2.plot(y,x,'g')
axes2.set_xlabel('y')
axes2.set_ylabel('x')
axes2.set_title('Figura Secundária')
```

##### Gráficos em paralelo

```python
# Dados
x = linspace(0,5,10)
y = x ** 2

# Divide a área de plotagem em dois subplots

fig, axes = plt.subplots(nrows = 1, ncols = 2)

# Loop pelos eixos para criar cada plot
for ax in axes:
    ax.plot(x,y,'r')
    ax.set_xlabel('x')
    ax.set_ylabel('y')
    ax.set_title('Título')

# Ajusta o layout
fig.tight_layout()
```

##### Gráfico de linhas com diferentes escalas

```python
# Dados

x = linspace(0,5,10)
y = x **2

# Cria os subplots
fig, axes = plt.subplots(1,2, figsize=(10,4))

# Cria o plot 1
axes[0].plot(x,x**2,x,exp(x))
axes[0].set_title('Escala Padrão')

# Cria o plot2
axes[1].plot(x,x**2,x,exp(x))
axes[1].set_yscale('log')
axes[1].set_title('Escala Logarítmica (y)')
```

```python
# Grid

# Dados

x = linspace(0,5,10)
y = x **2

# Cria os subplots
fig, axes = plt.subplots(1,2, figsize=(10,4))

# Grid padrão
axes[0].plot(x,x**2,x,x**3, lw =2)
axes[0].grid(True)

# Grid customizado
axes[1].plot(x,x**2,x,x**3, lw =2)
axes[1].grid(color = 'b', alpha = 0.7, linestyle = 'dashed', linewidth = 0.8)
```

```python
# Violin plot
sea.violinplot(df.idade,color='cyan')
```

# 3. Criando um gráfico estatístico com Matplotlib, Seaborn, Numpy e Pandas

```python
# Valores randômicos
np.random.seed(42)
n = 1000
pct_smokers = 0.2

# Variáveis
flag_fumantes = np.random.rand(n) < pct_smokers # O sinal de menor é que faz com que seja verdadeiro ou falso
idade = np.random.normal(40,10,n)
altura = np.random.normal(170,10,n)
peso = np.random.normal(70,10,n)

# Criando dataframe no Pandas
dados = pd.DataFrame({'altura':altura,'peso':peso,'flag_fumantes':flag_fumantes})

# Cria os dados para a variável flag_fumante
dados['flag_fumantes'] = dados['flag_fumantes'].map({True: 'Fumante',False:'Não fumante'})
```

```python
dados.shape
```

```python
dados.head()
```

```python
# Estilo do gráfico (Seaborne)
sea.set(style='ticks')

# lmplot
sea.lmplot(x = 'altura',
           y = 'peso',
           data = dados,
           hue = 'flag_fumantes',
           palette = ['tab:blue', 'tab:orange'],
           height = 7)

# Labels e título
plt.xlabel('Altura (cm)')
plt.ylabel('Peso (kg)')
plt.title('Relação entre altura e peso de fumantes e não fumantes')

# Remove as bordas
sea.despine()

plt.show()
```
